````python
# Importing the dataset
import pandas as pd
df = pd.read_csv("data/airline_data.csv")
df.head()
````

### Get Feature Type
if you want to know the type of feature ("Numerical", or "Categorical") of a particular column in your dataset, you can use the following method.

```bash
sdk.StatisticalTestRunner.get_feature_type(dataframe, column_name)

````python
column_1 = "airline"
print("""Data type of feature "{}" is {}""".format(column_1, sdk.StatisticalTestRunner.get_feature_type(df, column_1)))

column_2 = "time_taken_minutes"
print("""Data type of feature "{}" is {}""".format(column_2, sdk.StatisticalTestRunner.get_feature_type(df, column_2)))
````

### Executing Statistical test

if you want to automatically determine and execute the statistical test, you can use the following method.

```bash
sdk.StatisticalTestRunner.execute_statistical_test(dataframe, feature_1, feature_2, detailed = True, handle_missing = "remove")

````python
sdk.StatisticalTestRunner.execute_statistical_test(df, "price", "from", detailed = True, handle_missing = "remove")
````
